This Notbook plots the Cactus detected CMEs on top of particle data.

Setting the absolute path at the Scripts directory.

```python
import sys
import os
sys.path.append(os.path.abspath('../Scripts'))
```

loading up the methods to load and plot the data and CMEs.

```python
from load_plot_blk_data import load_blk_variables
from load_plot_cactus_data import parse_cactus_file
from plot_parameters_vs_cmes import plot
```

Loading in the data from the CDF files.

```python
trainingData = load_blk_variables('../data/Training_data/*.cdf')
untrainedData = load_blk_variables('../data/Untrained_data/*.cdf')
```

Merging both the dictionaries using numpy libraries.

```python
import numpy as np
data = {}
for key in trainingData:
    data[key] = np.concatenate([trainingData[key], untrainedData[key]])
```

Loading in the data from the Cactus text files.

```python
cme_times = (parse_cactus_file('../data/cmesept2024.txt') + 
             parse_cactus_file('../data/cmeoct2024.txt') + 
             parse_cactus_file('../data/cmenov2024.txt') +
             parse_cactus_file('../data/cmedec2024.txt') +
             parse_cactus_file('../data/cmejan2025.txt') +
             parse_cactus_file('../data/cmefeb2025.txt') +
             parse_cactus_file('../data/cmemar2025.txt') +
             parse_cactus_file('../data/cmeapril2025.txt')
            )
```

Generating the graphs:

```python
plot(data, cme_times)
```
