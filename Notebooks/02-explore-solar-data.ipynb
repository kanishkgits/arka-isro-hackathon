This notebook demonstrate the extraction of data from CDF files and plotting them(scaled).

Setting the absolute path at the root directory of the repo.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))
```

Importing the required methods from bulk handler script.

```python
from Scripts.load_plot_blk_data import plot_blk_parameters, load_blk_variables
```

Loading in the data from the CDF files.

```python
trainingData = load_blk_variables('../data/Training_data/*.cdf')
untrainedData = load_blk_variables('../data/Untrained_data/*.cdf')
```

Merging both the dictionaries using numpy libraries.

```python
import numpy as np
allData = {}
for key in trainingData:
    allData[key] = np.concatenate([trainingData[key], untrainedData[key]])
```

Generating the graphs of bulk parameters:

```python
plot_blk_parameters(allData)
```
