```python
import matplotlib.pyplot as plt
```

```python
import sys
import os
sys.path.append(os.path.abspath('../Scripts'))

from Bulkparameters import load_blk_variables
```

```python
DATA_PATH = '../data/BLK/*.cdf'
INVALID = -1e31
```

```python
data = load_blk_variables(DATA_PATH, INVALID)
```

```python
plt.figure(figsize=(20, 5)) 
plt.plot(data['time'], data['proton_bulk_speed'], label='Proton Bulk Speed', linewidth = 0.5)
plt.xlabel("Time")
plt.ylabel("Speed [km/s]")
plt.show()

plt.figure(figsize=(20, 5)) 
plt.plot(data['time'], data['alpha_bulk_speed'], label='Alpha Bulk Speed', linewidth = 0.5, color='tab:orange')
plt.xlabel("Time")
plt.ylabel("Speed [km/s]")
plt.show()
```

```python
plt.figure(figsize=(20, 5)) 
plt.plot(data['time'], data['proton_density'], label='Proton Density', linewidth = 0.5)
plt.xlabel("Time")
plt.ylabel("Number Density [#/cm]")
plt.show()


plt.figure(figsize=(20, 5)) 
plt.plot(data['time'], data['alpha_density'], label='Alpha Density', linewidth = 0.5, color='tab:orange')
plt.xlabel("Time")
plt.ylabel("Number Density [#/cm]")
plt.show()
```

```python
plt.figure(figsize=(20, 5)) 
plt.plot(data['time'], data['proton_thermal'], label='Proton Thermal', linewidth = 0.5)
plt.xlabel("Time")
plt.ylabel("Thermal speed [Km/sec]")
plt.show()

plt.figure(figsize=(20, 5))
plt.plot(data['time'], data['alpha_thermal'], label='Alpha Thermal', linewidth = 0.5, color='tab:orange')
plt.xlabel("Time")
plt.ylabel("Thermal speed [Km/sec]")
plt.show()
```

```python
plt.figure(figsize=(20, 5)) 
plt.plot(data['time'], data['proton_xvelocity'], label='Proton X-velocity', linewidth = 0.5)
plt.xlabel("Time")
plt.ylabel("Speed [km/s]")
plt.show()

plt.figure(figsize=(20, 5)) 
plt.plot(data['time'], data['proton_yvelocity'], label='Proton Y-velocity', linewidth = 0.5)
plt.xlabel("Time")
plt.ylabel("Speed [km/s]")
plt.show()

plt.figure(figsize=(20, 5)) 
plt.plot(data['time'], data['proton_zvelocity'], label='Proton Z-velocity', linewidth = 0.5)
plt.xlabel("Time")
plt.ylabel("Speed [km/s]")
plt.show()
```
